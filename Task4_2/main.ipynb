```python
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Import data
df = pd.read_csv('shrink_ray_dataset.csv')
print(df.shape)

# Define features (knob positions) and target (shrinkage)
X = df[['Power']]
Y = df['Shrinkage']
```

```python
# Create Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X, Y)
```

```python
# Desired shrinkage percentage
desired_shrinkage = 85

# Calculate the knob position for the desired shrinkage using the model's inverse prediction
knob_position_predicted = model.predict([[desired_shrinkage]])
print(f"Predicted knob position for {desired_shrinkage}% shrinkage: {knob_position_predicted[0]:.2f} MW")
```

```python
# Scatter plot
plt.scatter(X, Y, label='Data')
plt.xlabel('Power')
plt.ylabel('Shrinkage (%)')

# Linear regression line
plt.plot(X, model.predict(X), color='red', label='Linear Model')
plt.legend()

plt.title('Power vs. Shrinkage')
plt.show()
```

```python
from sklearn.metrics import r2_score

# Make predictions
Y_pred = model.predict(X)

# Calculate R-squared
r_squared = r2_score(Y, Y_pred)
print(f"R-squared: {r_squared:.4f}")
```
